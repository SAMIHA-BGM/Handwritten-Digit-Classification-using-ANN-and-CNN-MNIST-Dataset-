```python
#1- importing lib
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Input
```

```python
#2 data load n preprocess
text = open('/shakespear.txt', 'r').read().lower()
chars = sorted(set(text))
char_to_index = {char: index for index, char in enumerate(chars)}
index_to_char = {index: char for index, char in enumerate(chars)}
```

```python
#3 ip op seq
seq_length = 100
X = []
y = []
for i in range(0, len(text) - seq_length, seq_length):
    seq_in = text[i:i+seq_length]
    seq_out = text[i+seq_length]
    X.append([char_to_index[char] for char in seq_in])
    y.append(char_to_index[seq_out])

X = tf.keras.utils.to_categorical(X, num_classes=len(chars))
y = tf.keras.utils.to_categorical(y, num_classes=len(chars))
```

```python
#4 modeldef
model = Sequential()
model.add(Input(shape=(X.shape[1], X.shape[2])))
model.add(SimpleRNN(128, return_sequences=False))
model.add(Dense(len(chars), activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.001))
```

```python
#5 training
model.fit(X, y, batch_size=128, epochs=20)
```

```python
def generate_text(model, start_text, length=100):
    text = start_text
    for i in range(length):
        seq_in = [char_to_index[char] for char in text[-seq_length:]]
        seq_in = tf.keras.utils.to_categorical([seq_in], num_classes=len(chars))
        prediction = model.predict(seq_in, verbose=0)
        predicted_index = prediction.argmax()
        predicted_char = index_to_char[predicted_index]
        text += predicted_char
    return text
```

```python
start_text = "shall i compare thee to a summer's day"
generated_text = generate_text(model, start_text)
print(generated_text)
```
